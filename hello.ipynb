```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# -------------------------------
# CONFIGURATION
# -------------------------------
num_records = 3000

issue_types = ['Network', 'Software', 'Hardware', 'Security']
priorities = ['Low', 'Medium', 'High', 'Critical']
statuses = ['Closed', 'Open', 'In Progress']
agents = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Hank']

# SLA limits (in hours)
sla_limits = {'Critical': 4, 'High': 12, 'Medium': 24, 'Low': 48}

data = []

start_date = datetime(2025, 1, 1)

# -------------------------------
# DATA GENERATION LOOP
# -------------------------------
for i in range(num_records):

    ticket_id = f"T{1000 + i}"

    # Realistic weighted distributions
    issue = random.choices(issue_types, weights=[35, 40, 15, 10])[0]
    priority = random.choices(priorities, weights=[15, 50, 25, 10])[0]
    status = random.choices(statuses, weights=[75, 15, 10])[0]

    agent = random.choice(agents)

    # Random date within 6 months
    created_date = start_date + timedelta(
        days=random.randint(0, 180),
        hours=random.randint(0, 23)
    )

    # Time intelligence columns
    day = created_date.strftime("%A")
    month = created_date.strftime("%b")
    month_num = created_date.month

    # -------------------------------
    # CLOSED TICKET LOGIC
    # -------------------------------
    if status == 'Closed':

        if priority == 'Critical':
            res_time = random.uniform(1, 5)
        elif priority == 'High':
            res_time = random.uniform(4, 15)
        elif priority == 'Medium':
            res_time = random.uniform(12, 30)
        else:
            res_time = random.uniform(24, 72)

        # Round resolution time
        res_time = round(res_time, 2)

        resolved_date = created_date + timedelta(hours=res_time)

        # SLA check
        sla_status = 'Met' if res_time <= sla_limits[priority] else 'Not Met'

        # CSAT logic
        if sla_status == 'Met':
            csat = random.randint(4, 5)
        else:
            csat = random.randint(1, 3)

    # -------------------------------
    # OPEN / IN PROGRESS LOGIC
    # -------------------------------
    else:
        resolved_date = pd.NaT
        res_time = np.nan
        sla_status = np.nan
        csat = np.nan

    # Append row
    data.append([
        ticket_id, issue, priority, status,
        created_date, resolved_date, res_time,
        agent, sla_status, csat,
        day, month, month_num
    ])

# -------------------------------
# CREATE DATAFRAME
# -------------------------------
columns = [
    'Ticket ID', 'Issue Type', 'Priority', 'Status',
    'Created Date', 'Resolved Date', 'Resolution Time',
    'Assigned Agent', 'SLA Status', 'CSAT',
    'Day', 'Month', 'Month Number'
]

df = pd.DataFrame(data, columns=columns)

# -------------------------------
# EXPORT CSV
# -------------------------------
df.to_csv('IT_Support_Tickets.csv', index=False)

print("✅ Dataset generated successfully with", num_records, "records.")
```
